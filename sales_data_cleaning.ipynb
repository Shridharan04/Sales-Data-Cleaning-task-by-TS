```python
import pandas as pd
#loading the data set 
df = pd.read_csv("train.csv")
#seeing the first 5 rows
df.head()
```

```python
# how many rows and columns?
print("Shape:", df.shape)

#Column names, data types, missing values
df.info()
```

```python
df.isnull().sum()
```

```python
df[df['Postal Code'].isnull()]
```

```python
#Filling missing values
df['Postal Code'] = df['Postal Code'].fillna(5401)

```

```python
#verifying the work is done or not 
print(df['Postal Code'].isnull().sum())
```

```python
#to convert order date and ship date from text to proper date
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst = True)
df['Ship Date'] = pd.to_datetime(df['Ship Date'], dayfirst = True)
```

```python
#verifying the change 
print(df[['Order Date', 'Ship Date']].dtypes)
```

```python
df['Shipping Days'] = (df['Ship Date'] - df['Order Date']).dt.days
df[['Order Date', 'Ship Date', 'Shipping Days']].head(200)
```

```python
pd.set_option('display.max_rows', 60)
```

```python
#droping unnecessary columns
df = df.drop(columns = ['Row ID'])
```

```python
#verifying the work is done or not 
print(df.columns.tolist())
```

```python
#Saving the cleaned dataset as a new file
df.to_csv("train_cleaned.csv", index = False)

print("File Saved Successfully")
```
